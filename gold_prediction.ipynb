# Step 1: Loading Data...

```python
import yfinance as yf
import datetime
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
end_date = datetime.datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.datetime.today() - datetime.timedelta(days=math.ceil(20*365.25))).strftime('%Y-%m-%d')
ticker = 'GC=F'
data = yf.Ticker(ticker).history(start=start_date, end=end_date)
data.reset_index(inplace=True)
#Chuyển đổi định dạng dữ liệu
data['Date'] = pd.to_datetime(data['Date'], format="%Y-%m-%d %H:%M:%S%z")

# Chuyển đổi sang định dạng ngày mà không có thời gian và múi giờ
data['Date'] = data['Date'].dt.date
```

# Step 2: Visualization

```python
#Biểu đồ thời gian cho giá đóng cửa
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')
plt.title('Xu hướng giá đóng cửa của Vàng theo thời gian')
plt.xlabel('Thời gian')
plt.ylabel('Giá đóng cửa (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Phân bố giá đóng cửa

plt.figure(figsize=(8, 6))
sns.histplot(data['Close'], bins=30, kde=True, color='purple')
plt.title('Phân bố giá đóng cửa của Vàng')
plt.xlabel('Giá đóng cửa (USD)')
plt.ylabel('Tần suất')
plt.show()
```

```python
# Biểu đồ khối lượng giao dịch theo thời gian
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Volume'], label='Khối lượng giao dịch', color='orange')
plt.title('Khối lượng giao dịch của vàng theo thời gian')
plt.xlabel('Thời gian')
plt.ylabel('Khối lượng giao dịch(kUSD)')
plt.legend()
plt.grid()
plt.show()
```
